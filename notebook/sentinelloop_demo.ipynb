## 3. Run a Triage Scenario

Let's simulate a SOC analyst requesting the triage of a suspicious IP address.

```python
def run_triage(indicator):
    print(f"🔍 Initiating Triage for: {indicator}\n")
    print("🤖 [Validator Agent] Classifying indicator and setting priority...")
    print("🕵️‍♂️ [Recon Agent] Gathering OSINT data via VexScan MCP...")
    print("🔬 [Analysis Agent] Checking for required forensic analysis via Phexor MCP...")
    print("📝 [Report Agent] Compiling final structured report...\n")
    
    try:
        # Execute the ADK Sequential Agent
        result = root_agent(f"Investigate this security indicator: {indicator}")
        print("="*50)
        print("📄 FINAL SOC TRIAGE REPORT")
        print("="*50)
        print(result)
    except Exception as e:
        print(f"Error during execution: {e}")

# Example 1: Known Malicious IP
run_triage("45.33.32.156")
```

## 4. Batch IOC Scanning

SentinelLoop can also handle batches of Indicators of Compromise (IOCs) simultaneously using our custom `batch_ioc_scan` MCP tool.

```python
iocs = [
    "45.33.32.156",
    "malware.wicar.org",
    "8.8.8.8" # Control (Safe)
]

print(f"🔍 Initiating Batch Triage for {len(iocs)} indicators...")
# In a full environment, the agent would use the batch_ioc_scan tool automatically.
try:
    result = root_agent(f"Investigate this batch of indicators: {', '.join(iocs)}")
    print("\n" + "="*50)
    print("📄 BATCH SOC TRIAGE REPORT")
    print("="*50)
    print(result)
except Exception as e:
    print(f"Error during execution: {e}")
```
